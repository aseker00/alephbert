```python
import pandas as pd
from bclm import treebank as tb
from transformers import AutoTokenizer
```

```python
partition = tb.spmrl('data/raw')
```

```python
roberta_tokenizer = AutoTokenizer.from_pretrained("./experiments/transformers/roberta-bpe-byte-v1")
bert_tokenizer = AutoTokenizer.from_pretrained("./experiments/transformers/bert-wordpiece-v1")
```

```python
dev = partition['dev']
```

```python
df = dev[1]
df
```

```python
raw_input_df = df[['token_id', 'token']].drop_duplicates()
output_seq_df = df[['token_id', 'form']]
```

```python
output_seq_df
```

```python
roberta_input_seq_df = pd.DataFrame([(row.token, row.token_id, xt) for index, row in raw_input_df.iterrows() for xt in roberta_tokenizer.tokenize(row.token)], columns=['token','token_id','xtoken'])
bert_input_seq_df = pd.DataFrame([(row.token, row.token_id, xt) for index, row in raw_input_df.iterrows() for xt in bert_tokenizer.tokenize(row.token)], columns=['token','token_id','xtoken'])
```
